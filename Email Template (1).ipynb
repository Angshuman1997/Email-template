## Extracting Materials

##### Fetching or extracting links, phone nimbers,email id from email body(if present) of each individual and creating  dataframe

```python
df1=df['Email Body']
l=[]
l1=[]
d=[]
d1=[]
d2=[]
e=[]
f1=[]
g=[]
l.append(df['Email Body'].values)
l1.append(df['first_name'].values)
for i in range(len(df1)):
    s=re.findall(r"\+?\d{1,3}?[- .]?\(?(?:\d{2,3})\)?[- .]?\d\d\d[- .]?\d\d\d\d",df1[i])
    st=re.findall('\S+@\S+',df1[i])
    lin=re.findall('\S+//\S+',df1[i])
    ph= re.findall(r'\b[0123456789]\d{9}\b',df1[i], flags=0)
    if len(lin)==0:
        d.append('No')
        g.append(" - ")
    else:
        d.append('Yes')
        g.append(lin)
    if len(st)==0:
        d1.append('No')
        e.append(" - ")
    else:
        d1.append('Yes')
        e.append(st)
    if len(s)==0:
        d2.append('No')
        f1.append(" - ")
    else:
        d2.append('Yes')
        f1.append(s)
        
print(len(f1))
df3=pd.DataFrame({'First Name':l1[0],'Email Body':l[0],'Fetchable Link':d,'Fetched Link':g,'Fetchable Email-ID':d1,'Fetched Email-ID':e,'Fetchable Phno':d2,'Fetched Ph':f1})
df3 
```
